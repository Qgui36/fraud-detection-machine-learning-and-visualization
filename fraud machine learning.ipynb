# 2. data cleaning

```python
data_ml=data.drop(['nameOrig','nameDest','isFlaggedFraud'],axis=1)
```

```python
data_ml.columns
```

```python
data_ml_tran_cash=data_ml[(data_ml['type']=='TRANSFER')|(data_ml['type']=='CASH_OUT')]
```

```python
# get dummy variable for the types
data_ml_dummy=pd.get_dummies(data_ml_tran_cash)
```

```python
data_ml_dummy.head()
```

```python
X = data_ml_dummy.drop(['isFraud'],axis=1)
y = data_ml_dummy['isFraud']
```

```python
# skip this step will lower the score to 0.9993286192296447
X['errorBalanceOrig'] = X.newbalanceOrig + X.amount - X.oldbalanceOrg
X['errorBalanceDest'] = X.oldbalanceDest + X.amount - X.newbalanceDest
```

```python
# Long computation in this cell (~2.5 minutes)
x_3d = 'errorBalanceDest'
y_3d = 'step'
z_3d = 'errorBalanceOrig'
zOffset = 0.02
limit = len(X)

sns.reset_orig() # prevent seaborn from over-riding mplot3d defaults

fig = plt.figure(figsize = (10, 12))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X.loc[y == 0, x_3d][:limit], X.loc[y == 0, y_3d][:limit], \
  -np.log10(X.loc[y == 0, z_3d][:limit] + zOffset), c = 'g', marker = '.', \
  s = 1, label = 'genuine')
    
ax.scatter(X.loc[y == 1, x_3d][:limit], X.loc[y == 1, y_3d][:limit], \
  -np.log10(X.loc[y == 1, z_3d][:limit] + zOffset), c = 'r', marker = '.', \
  s = 1, label = 'fraudulent')

ax.set_xlabel(x_3d, size = 16); 
ax.set_ylabel(y_3d + ' [hour]', size = 16); 
ax.set_zlabel('- log$_{10}$ (' + z_3d + ')', size = 16)
ax.set_title('Error-based features separate out genuine and fraudulent \
transactions', size = 20)

plt.axis('tight')
ax.grid(1)

noFraudMarker = mlines.Line2D([], [], linewidth = 0, color='g', marker='.',
                          markersize = 10, label='genuine')
fraudMarker = mlines.Line2D([], [], linewidth = 0, color='r', marker='.',
                          markersize = 10, label='fraudulent')

plt.legend(handles = [noFraudMarker, fraudMarker], \
           bbox_to_anchor = (1.20, 0.38 ), frameon = False, prop={'size': 16});
```

```python
Xfraud = X.loc[y == 1] # update Xfraud & XnonFraud with cleaned data
XnonFraud = X.loc[y == 0]
                  
correlationNonFraud = XnonFraud.loc[:, X.columns != 'step'].corr()
mask = np.zeros_like(correlationNonFraud)
indices = np.triu_indices_from(correlationNonFraud)
mask[indices] = True

grid_kws = {"width_ratios": (.9, .9, .05), "wspace": 0.2}
f, (ax1, ax2, cbar_ax) = plt.subplots(1, 3, gridspec_kw=grid_kws, \
                                     figsize = (14, 9))

cmap = sns.diverging_palette(220, 8, as_cmap=True)
ax1 =sns.heatmap(correlationNonFraud, ax = ax1, vmin = -1, vmax = 1, \
    cmap = cmap, square = False, linewidths = 0.5, mask = mask, cbar = False)
ax1.set_xticklabels(ax1.get_xticklabels(), size = 16); 
ax1.set_yticklabels(ax1.get_yticklabels(), size = 16); 
ax1.set_title('Genuine \n transactions', size = 20)

correlationFraud = Xfraud.loc[:, X.columns != 'step'].corr()
ax2 = sns.heatmap(correlationFraud, vmin = -1, vmax = 1, cmap = cmap, \
 ax = ax2, square = False, linewidths = 0.5, mask = mask, yticklabels = False, \
    cbar_ax = cbar_ax, cbar_kws={'orientation': 'vertical', \
                                 'ticks': [-1, -0.5, 0, 0.5, 1]})
ax2.set_xticklabels(ax2.get_xticklabels(), size = 16); 
ax2.set_title('Fraudulent \n transactions', size = 20);

cbar_ax.set_yticklabels(cbar_ax.get_yticklabels(), size = 14);
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 42)
```

```python
# takes 30 min to run, score = 0.999980147342812
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sort_import=pd.DataFrame(sorted(zip(rf.feature_importances_, X.columns), reverse=True))
```

```python
sns.barplot(x=sort_import[0], y=sort_import[1])
plt.xlabel("feature importances")
plt.ylabel("features")
```

```python
# for tree visualization, set max_depth = 3
# takes 30 min to run, score = 0.998197017769933
rf = RandomForestClassifier(max_depth = 3,n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
estimator = rf.estimators_[5]
export_graphviz(estimator, out_file='fraud_tree.dot', feature_names = X_train.columns,
                class_names = ['Genuine','Fraudulent'], rounded = True, proportion = False, precision = 2, 
                filled = True)
```
